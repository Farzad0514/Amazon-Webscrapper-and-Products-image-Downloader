```python
def getImages(url, fileName):
    import urllib.request
    from bs4 import BeautifulSoup
    import json
    import time
    
    #url= 'https://www.amazon.com.au/Thermos-Stainless-Ounce-Travel-Handle/dp/B00ETP7DAW/ref=sr_1_103?dchild=1&m=ANEGB3WVEVKZB&qid=1601494762&refinements=p_36%3A3900-5000%2Cp_n_free_shipping_eligible%3A5363790051%2Cp_72%3A2547912051%2Cp_6%3AANEGB3WVEVKZB&rnid=5355409051&s=kitchen&sr=1-103'
    newReq= urllib.request.urlopen(url)
    print(newReq.getcode())
    
    time.sleep(10)
    
    soup= BeautifulSoup(newReq,  "html.parser")
    all=soup.find_all("script")

    allstr=str(all)
    ImageBlockATF= allstr.index("ImageBlockATF")
    AboveTheFold= allstr.index("P.AboveTheFold")
    data=allstr[int(ImageBlockATF):int(AboveTheFold)]
    initial= data.index('initial')
    colorToAsin= data.index('colorToAsin')
    raw = data[int(initial)-3:int(colorToAsin)-3]
    modi_raw=raw.replace('\n', "").replace("'", '"')

    jsonData=json.loads(modi_raw)
    test=jsonData['initial']

    HQ_images=[]
    for k in range(0, len(test)):
        allImages = test[k]
        hiRes= allImages['hiRes']
        HQ_images.append(hiRes)


    for i in range (0, int(len(HQ_images))):

        titlePrefix= str(i+1)+str("_")
        try:
            print("Image number_" + str(i+1))
            print("Downloading starts...\n")
            urllib.request.urlretrieve(str(HQ_images[i]), str(titlePrefix) + str(fileName)+str('.jpg'))
            print(str((HQ_images[i])))
            print("Download completed..!!")
            print()
        except Exception as e:
            print(e)

    print("All files Downloaded")
    
    
    

# To download images in folders in a loop
#csvfilePath is the path of productUrl.csv file
#directory is the path of in which you want to store folders of images
def getImages_loop(csvfilePath, directory ):
    import pandas
    import time
    import os

    raw_dataUrl = pandas.read_csv(csvfilePath)
    dataUrl = raw_dataUrl['raw']
    os.chdir(directory)
    lenDataUrl= len(dataUrl)
    #for i in range (0, int(lenDataUrl)):
    for i in range (2, 30):

        fileName= 'prd_'+str(i+1)
        os.mkdir(fileName) # Make a folder
        os.chdir(fileName) #Change directory to new folder
        
        #print(len(dataUrl))
        url = dataUrl[i]
        # download images
        print(url)
        print(fileName)
        try:
            getImages(url, fileName)
        except:
            print('getImages failed for: '+str(i))
            pass

        #parent = os.path.abspath('..')
        
        
        os.chdir(directory) # change directory to original path

        print('download completed for fileName: ' + str(fileName))
        
        time.sleep(10)
```

```python
# directory= input('directory :')
# csvfilePath= input('csvfilePath :')
directory= 'D://Ebay//aus/oct//'
csvfilePath= 'D://Ebay//aus//oct//productsUrl.csv'
getImages_loop(csvfilePath, directory )
```
